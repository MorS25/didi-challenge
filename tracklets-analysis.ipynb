KITTI Dataset Analysis
===

```python
import os
import glob

where_search = 'tracklets'
tracklet_name = 'tracklet_labels.xml'

tracklets_files = []

for filename in glob.iglob(os.path.join(where_search, '**', tracklet_name), recursive=True):
    tracklets_files.append(filename)
```

```python
from tracklets import Tracklets

tracks = Tracklets.load(tracklets_files)
```

```python
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline

tracks.frame.plot.hist(alpha=0.4, bins=20)
```

```python
tracks[['h', 'w', 'l']].plot.hist(alpha=0.4, stacked=True, bins=20)
```

### Frequencies of X, Y, Z coordinates

```python
tracks[['tx', 'ty', 'tz']].plot.hist(alpha=0.3, stacked=False, bins=50, logy=True)
```

### Number of objects per frame

```python
tracks.groupby(['frame']).size().plot.hist(alpha=0.3, stacked=False, bins=50)
```
